## Linear Regression Single Variable

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn import linear_model
```

```python
df = pd.read_excel('homeprice.xlsx')
df
```

```python
plt.xlabel('area(sqr ft)')
plt.ylabel('price(US$)')
plt.scatter(df.area, df.price, color='red', marker='+')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']], df.price)
```

```python
reg.predict([[2600]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
135.78767123*3300+180616.43835616432
```

```python
d = pd.read_excel('area.xlsx')
d
```

```python
p = reg.predict(d)
```

```python
d['price'] = p
```

```python
d
```

```python
d.to_excel('area.xlsx', index=False)
```

```python
plt.xlabel('area', fontsize=20)
plt.ylabel('price', fontsize=20)
plt.scatter(df.area, df.price, color='red', marker='+')
plt.plot(df.area, reg.predict(df[['area']]), color='blue')
```

### Canada_Per_Capita_Income_ Price Prediction Linear Regression

```python
df1 = pd.read_csv('canada_per_capita_income.csv')
df1
```

```python
df1['price'] = df1['per capita income (US$)']
df1
```

```python
df2 = df1.drop(['per capita income (US$)'], axis='columns')
df2
```

```python
plt.xlabel('Year')
plt.ylabel('price(US$)')
plt.scatter(df2.year, df2.price, color='red', marker='+')
```

```python
reg1 = linear_model.LinearRegression()
reg1.fit(df2[['year']], df2.price)
```

```python
reg1.predict([[2023]])
```

```python
reg1.coef_
```

```python
reg1.intercept_
```

```python
828.46507522*2023+-1632210.7578554575
```

```python
d1 = pd.read_excel('year.xlsx')
d1
```

```python
p1 = reg1.predict(d1)
```

```python
d1['price'] = p1
d1
```

```python
d1.to_csv('canada_per_capita_income.csv', index=False)
```

```python
plt.xlabel('year', fontsize=20)
plt.ylabel('price', fontsize=20)
plt.scatter(d1.year, d1.price, color='red', marker='+')
plt.plot(d1.year, reg1.predict(d1[['year']]), color='blue')
```
